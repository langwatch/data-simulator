```python
# Step 2: Import necessary libraries
import os
from google.colab import files
import pandas as pd
from data_simulator import DataSimulator
from IPython.display import display, HTML
```

```python
# Step 3: Set up OpenAI API key
from google.colab import userdata
import getpass

api_key = getpass.getpass('Enter your OpenAI API key: ')
```

```python
# Step 4: Initialize the DataSimulator
generator = DataSimulator(api_key=api_key)
```

```python
# Step 5: Define a utility function to display results nicely
def display_results(results):
    """Display results in a nicely formatted pandas DataFrame."""
    # Extract relevant data
    data = [{
        'ID': item['id'],
        'Query': item['query'],
        'Document': item['document'][:100] + "..." if len(item['document']) > 100 else item['document'],
        'Answer': item['answer'][:150] + "..." if len(item['answer']) > 150 else item['answer']
    } for item in results]

    # Create DataFrame
    df = pd.DataFrame(data)

    # Print header
    print(f"\n=== Generated {len(results)} results ===\n")

    # Display the DataFrame with styling
    display(HTML(df.to_html(escape=False)))
```

```python
# Step 6: Generate data

results = generator.generate_from_docs(
    file_paths=["./data-simulator/test_data/nike_10k.pdf"],
    context="You're a financial support assistant for Nike, helping a financial analyst decide whether to invest in the stock.",
    example_queries="how much revenue did nike make last year\nwhat risks does nike face\nwhat are nike's top 3 priorities"
)

display_results(results)
```
